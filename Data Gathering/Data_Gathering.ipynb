```python
#slice kolom 
df_tanpaid = df.iloc[:,2:] #(2 kolom -> kolom unnamed dan id di slice)
df_tanpaid
```

```python
#jika assist terbanyak
df_rapi = df_tanpaid.sort_values(by="assists", ascending=False)
df_rapi.head(15)
```

```python
df_rapi = df_tanpaid.sort_values(by="player_name", ascending=True)
df_rapi.head(15) #15 data teratas setelah dirapikan
```

```python
#menunjukkan ringkasan statistik dari kolom-kolom (hanya numerik)
df_rapi.describe()
```

```python
#menunjukkan ringkasan statistik dari kolom-kolom (semua)
df_rapi.describe(include='all')
```

```python
#jika dataframe tanpaid
df_tanpaid.describe(include='all')
```

```python
#mencari rata-rata
df_rapi[['goals', 'games', 'assists', 'shots']].mean()
```

```python
#mencari modus
df_rapi[['team_title']].mode()
```
